Question 2 Print the sum of all prime number between A to B.


```python
a = int(input("Enter the value of A. "))
b = int(input("Enter the value of B."))
sum = 0
for n in range(a, b+1):
  if n == 1:
    continue
  for i in range(2, n):
    if n % i == 0:
      break

  else:
    sum += n
print(f"Sum of all Prime numbers between A and B is {sum}")
```

Question 5. Check if a string is a palindrome using loop.


```python
string = input("Enter the String. ")
left = 0
right = len(string) - 1
palindrome = True
while left < right:
  if string[left] != string[right]:
    palindrome = False
    break
  left += 1
  right -= 1
if palindrome:
  print(f"{string} is a palindrome.")
else:
  print(f"{string} is not a palindrome")
```

Question 7. Check if a number is Armstrong number.


```python
num = int(input("Enter a number. "))
num_len = len(str(num))
armstrong_sum = 0
for d in str(num):
  armstrong_sum += int(d) ** num_len

if armstrong_sum == num:
  print(f"The Given Number {num} is a Armstrong Number")
else:
  print(f"The Given Number {num} is not a Armstrong Number")
```

Q21. Converting a decimal number to binary, octal and hexadecimal involves dividing the decimal number by the base repeatedly and noting the remainders at each step.

```python
number = float(input("Enter a decimal number here......."))
temp = int(number)

# Decimal to binary
binary = ""
t = temp
while t > 0:
  binary = str(t%2)+binary
  t //= 2
print(f"The binary of the given decimal number {number} is " + binary)

# Decimal to octal
octal = ""
t = temp
while t > 0:
  octal = str(t % 8) + octal
  t //= 8
print(f"The octal of the given decimal number {number} is " + octal)

# Decimal to hexadecimal
hex_digits = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
hexadecimal = ""
t = temp
while t > 0:
  hexadecimal = hex_digits[t % 16] + hexadecimal
  t //= 16
print(f"The hexadecimal of the given decimal number {number} is " + hexadecimal)
```

Q28. WAP to check if given array is Monotonic.

```python
def monotonic(arr):
    increasing = True
    Decreasing = True

    for i in range(1, len(arr)):
        if arr[i] < arr[i - 1]:
            increasing = False
        if arr[i] > arr[i - 1]:
            Decreasing = False

    return increasing or Decreasing


arr = np.array(list(map(int, input("Enter array elements: ").split(","))))

if monotonic(arr):
    print("Array is Monotonic")
else:
    print("Array is NOT Monotonic")
```

Q64. Define a class with a generator which can iterate the numbers, which are divisible by 7, between a given range 0 and n.

```python
class divBySeven:
  def __init__(self, n):
    self.n = n

  def generate(self):
    for i in range(0, self.n+1):
      if i == 0:
        continue
      elif i % 7 == 0:
        yield i
n = int(input("Enter the value of n......"))
obj = divBySeven(n)
for num in obj.generate():
  print(num)
```

# Iris Dataset Modelling Using NumPy

```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target

print("X shape: ", X.shape)
print("y shape: ", y.shape)
print('Features name: ', iris.feature_names)
print('Class name: ', iris.target_names)
```

```python
# inspect the data
print(X[:5])
print(y[:5])
```

```python
# Basic Statistics
mean = np.mean(X, axis = 0)
std = np.std(X, axis=0)

for i, name in enumerate(iris.feature_names):
  print(name, "mean:", round(mean[i], 2), "Std: ",round(std[i], 2))

```

```python
# Normaize the features (Z-score)
X_norm = (X-mean) / std

print(X_norm[:5])
```

```python
# Train-Test Split (Manual)
# 80% training data
# 20% testing data

np.random.seed(42)
indices = np.random.permutation(len(X_norm))
split = int(0.8 * len(X_norm))

train_idx = indices[:split]
test_idx = indices[split:]

X_train, X_test = X_norm[train_idx], X_norm[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

print("Train samples: ", X_train.shape[0])
print("Test samples: ", X_test.shape[0])
```

```python
# Nearest Centroid Classifier

centroids = np.array([
    X_train[y_train==0].mean(axis=0),
    X_train[y_train == 1].mean(axis=0),
    X_train[y_train == 2].mean(axis=0)
])
print(centroids)
```

```python
# Prediction Function
def Predict (X, centroids):
  distances = np.linalg.norm(
      X[:, None, :] - centroids[None, :, :],
      axis=2
  )
  return np.argmin(distances, axis=1)
```

```python
y_pred = Predict(X_test, centroids)
accuracy = np.mean(y_pred == y_test)
print("Model Accuracy: ", accuracy)
```
